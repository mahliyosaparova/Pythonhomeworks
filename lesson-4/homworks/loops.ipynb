## Loops

```python
n = 10

while n <= 20:
    print("Executed. n =", n)
    n += 1

print("Outside loop")
```

```python
while True:
    print("Still running")
    break
```

```python
cnt = 1

while True:
    print("cnt=", cnt)
    cnt += 1
    if cnt == 10:
        break
```

Continue

```python
cnt = 1

while cnt < 20:
    cnt += 1
    if cnt in (10, 14, 16):
        continue
    print("cnt=", cnt)
```

```python
l = [1, 2, 3]
i = 0
check_number = 5
while i < len(l):
    if l[i] == check_number:
        print("Found")
        break
    i += 1
else:
    print("Number not found")
```

## For Loop

```python
my_list = [1, 2, 3]
other_list = []
i = 0
while i < len(my_list):
    a = my_list[i]
    other_list.append(a**2)
    i += 1

print(other_list)
```

```python
other_list = []
for a in my_list:
    other_list.append(a**2)

print(other_list)
```

```python
my_str = 'abcdefghi'
for char in my_str:
    print(char)
```
